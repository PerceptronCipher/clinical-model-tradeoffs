```python
# import important libraries
from ucimlrepo import fetch_ucirepo, list_available_datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# check datasets that can be imported
list_available_datasets()
```

The project demands we opt for the heart disease datasets

```python
# Fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets
```

```python
# metadata and inspection
print(f"Features Shape: {X.shape}\nTargets Shape: {y.shape}")
print(f"Features Information: {X.info()} \nTargets Information: {y.info()}")
print(f"Features: \n{X.head()} \nTargets: \n{y.head()}")
```

```python
# Check for missing values 
print(f"Number of Missing Values in Features: {X.isnull().sum()} \nNumber of Missing Values in Targets: {y.isnull().sum()}")
```

```python
# Check for Targets distribution
print(y.value_counts().sort_index())
```

```python
# Visualize distribution
y.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Heart Disease Level')
plt.ylabel('Number of Patients')
plt.title('Target Distribution')
plt.show()
```
